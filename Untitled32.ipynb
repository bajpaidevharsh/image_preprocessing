```python
import cv2
import numpy as np
```

```python
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("/content/chess_data_data_1.jpeg")
img=cv2.resize(img,(500,500))
```

```python
cv2_imshow(img)
```

**converting into hsv is optional**

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
cv2_imshow(img_hsv)
```

```python
def empty(x):
    pass
```

```python
cv2.namedWindow('track')
cv2.resizeWindow('track', 500, 500)
kernel = np.ones((5,5), np.uint8)
cv2.createTrackbar('hue_min', 'track', 0, 179, empty)
cv2.createTrackbar('hue_max', 'track', 179, 179, empty)
cv2.createTrackbar('sat_min', 'track', 0, 255, empty)
cv2.createTrackbar('sat_max', 'track', 255, 255, empty)
cv2.createTrackbar('value_min', 'track', 0, 255, empty)
cv2.createTrackbar('value_max', 'track', 255, 255, empty)
cv2.createTrackbar('erode', 'track', 0, 5, empty)
cv2.createTrackbar('dilate', 'track', 0, 5, empty)
cv2.createTrackbar('canny', 'track', 0, 15, empty)
```

**this is how the track will look like**

```python
cv2.imread('/content/track_bars.png')
```

```python

while True:
    hue_min = cv2.getTrackbarPos('hue_min', 'track')
    hue_max = cv2.getTrackbarPos('hue_max', 'track')
    sat_min = cv2.getTrackbarPos('sat_min', 'track')
    sat_max = cv2.getTrackbarPos('sat_max', 'track')
    value_min = cv2.getTrackbarPos('value_min', 'track')
    value_max = cv2.getTrackbarPos('value_max', 'track')
    erode = cv2.getTrackbarPos('erode', 'track')
    dilate = cv2.getTrackbarPos('dilate', 'track')
    canny = cv2.getTrackbarPos('canny', 'track')

    # Define the lower and upper HSV ranges based on trackbar positions
    low = np.array([hue_min, sat_min, value_min])
    high = np.array([hue_max, sat_max, value_max])

    # Create a mask based on the current HSV range
    mask = cv2.inRange(image_hsv, low, high)

    # Apply Canny edge detection
    canny_edge = cv2.Canny(mask, canny * 10, canny * 10)

    # Apply dilation and erosion
    if dilate > 0:
        canny_edge = cv2.dilate(canny_edge, kernel, iterations=dilate)
    if erode > 0:
        canny_edge = cv2.erode(canny_edge, kernel, iterations=erode)

    # Display the mask
    cv2.imshow('output', canny_edge)

    # Break the loop when the 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cv2.destroyAllWindows()
```
